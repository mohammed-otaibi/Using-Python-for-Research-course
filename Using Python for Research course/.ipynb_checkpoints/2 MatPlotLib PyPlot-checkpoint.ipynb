# Customizing Your Plots:<br><br>

xlabel(), ylabel(),<br>
savefig().<br>   

```python
x= np.linspace(0,10,20)
y1= x** 2.0
y2= x** 1.5
```

```python
plt.plot(x, y1, 'bo-', linewidth= 2, markersize= 12);
plt.plot(x, y2, 'gs-', linewidth= 2, markersize= 12);
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.plot(x, y1, 'bo-', linewidth= 2, markersize= 12);
plt.plot(x, y2, 'gs-', linewidth= 2, markersize= 12);
plt.xlabel("X")
plt.ylabel("Y")
#plt.axis([xmin, xmax, ymin, ymax])
plt.axis([0.5,11,-5,105])
```

```python
plt.plot(x, y1, 'bo-', linewidth= 2, markersize= 12, label="First");
plt.plot(x, y2, 'gs-', linewidth= 2, markersize= 12, label="Second");
plt.xlabel("X")
plt.ylabel("Y")
#plt.axis([xmin, xmax, ymin, ymax])
plt.axis([0.5,11,-5,105])
plt.legend(loc="upper left")
```

```python
plt.plot(x, y1, 'bo-', linewidth= 2, markersize= 12, label="First");
plt.plot(x, y2, 'gs-', linewidth= 2, markersize= 12, label="Second");
plt.xlabel("X")
plt.ylabel("Y")
#plt.axis([xmin, xmax, ymin, ymax])
plt.axis([0.5,11,-5,105])
plt.legend(loc="upper left")
# to save a plot as .pdf use savefig()
plt.savefig("Myplot.pdf")
```

# Plotting Using Logarithmic Axes

<br>
In some plots, it's helpful to have one or both axes be logarithmic.<br>
This means that for any given point to be plotted, its x or y-coordinate,<br>
or both, are transformed using the log function.<br>

So what ends up being plotted on the x-axes <br>
is not the original x-coordinate, but instead the value of log x <br>
The logarithm is taken by default in base 10 :<bR><BR>

```python
x= np.linspace(0,10,20)
y1= x**2.0
y2= x**1.5
plt.loglog(x, y1, 'bo-', linewidth= 2, markersize= 12, label="First")
plt.loglog(x, y2, 'gs-', linewidth= 2, markersize= 12, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
```

```python
x= np.logspace(-1,1,40)
y1= x**2.0
y2= x**1.5
plt.loglog(x, y1, 'bo-', linewidth= 2, markersize= 12, label="First")
plt.loglog(x, y2, 'gs-', linewidth= 2, markersize= 12, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
```

```python
x = np.logspace(0,1,10)
y = x**2
plt.loglog(x,y,"bo-")
```

# Gamma Distrebuition

```python
x= np.random.gamma(2,3,100000)
```

```python
plt.hist(x, bins=30);
```

```python
plt.hist(x, bins=30, normed=True);
```

```python
plt.hist(x, bins=30, cumulative=True);
```

```python
plt.hist(x, bins=30, cumulative=True, normed=True);
```

```python
plt.hist(x, bins=30, cumulative=True, normed=True, histtype='step');
```

# to compention all the last 4 hist in one plot 

```python
plt.figure()
plt.subplot(221)
plt.hist(x, bins=30);
plt.subplot(222)
plt.hist(x, bins=30, normed=True);
plt.subplot(223)
plt.hist(x, bins=30, cumulative=True);
plt.subplot(224)
plt.hist(x, bins=30, cumulative=True, normed=True, histtype="step");
```
