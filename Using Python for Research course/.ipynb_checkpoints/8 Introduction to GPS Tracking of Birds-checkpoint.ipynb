# Simple Data Visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ix = birddata.bird_name == "Eric"
z= birddata[ix][['latitude','longitude']]
x= z['longitude']
y= z['latitude']
```

```python
plt.figure(figsize=(7 ,7))
plt.plot(x,y, 'b.');

# In this plot you can see the migration path that Eric has taken.
# Again, the flight trajectory will be substantially distorted
# because we did not implement any type of cartographic projection.
# Nevertheless, we get a decent sense of the flight path.
```

```python
# print unique bird name.
bird_names= pd.unique(birddata["bird_name"])
bird_names
```

```python
plt.figure(figsize=(7,7))
for bird_name in bird_names:
    xi= birddata['bird_name']==bird_name
    x, y= birddata.longitude[xi], birddata.latitude[xi]
    plt.plot(x, y, ".", label= bird_name)
plt.title("the trajectories for all three birds")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.legend(loc="lower right");
```

```python
ix= birddata["bird_name"] == "Eric"
x= birddata["speed_2d"][ix]
nas= np.isnan(x)
plt.hist(x[~nas], np.linspace(0, 30, 20),normed=True, label= "Eric")
plt.xlabel("2D speed (m/s)")
plt.ylabel("Freq")
plt.legend(loc= "upper right")
```

```python
birds_name=  birddata["bird_name"].unique()
for name in birds_name:
    ix= birddata["bird_name"] == name
    x= birddata["speed_2d"][ix]
    nas= np.isnan(x)
    plt.hist(x[~nas], np.linspace(0, 30, 20),normed=True, alpha=0.5, label= name)
plt.xlabel("2D speed (m/s)")
plt.ylabel("Freq")
plt.legend(loc= "upper right")
```

```python
time_stamp=[]
for k in range(birddata.shape[0]):
    time_stamp.append(datetime.datetime.strptime(birddata['date_time'][k][:-3], "%Y-%m-%d %H:%M:%S"))
(time_stamp[2])
```

```python
birddata["timestamp"]= time_stamp
```

```python
birddata.head()
```

```python
type(birddata['timestamp'][0])
```

#### Let's now practice doing arithmetic with our timestamps.

```python
birddata['timestamp'][4] - birddata['timestamp'][3]
```

```python
elapsed_time=[]
ix= birddata["bird_name"]== "Eric"
times=birddata['timestamp'][ix]
for time in times:
    elapsed_time.append(time - times[0])
```

```python
elapsed_time[1000]
```

#### But how can we measure time in certain units, like hours or days?

```python
elapsed_time[1000] / datetime.timedelta(days=1)
# I create a time delta object that's equal to 1 day.
# I can accomplish that by using the key word days equals to 1.
# In this case, I know that exactly 12.084 and so on days.
```

```python
elapsed_time[1000] / datetime.timedelta(hours=1)
```

```python
plt.plot(np.array(elapsed_time) / datetime.timedelta(days= 1))
plt.xlabel('Observation')
plt.ylabel('Elapsed time (days)')
```

# Using the Cartopy Library

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python

proj = ccrs.Mercator() 
plt.figure(figsize=(10,10)) 
ax = plt.axes(projection=proj) 
ax.set_extent((-25.0, 20.0, 52.0, 10.0)) 
ax.add_feature(cfeature.LAND) 
ax.add_feature(cfeature.OCEAN) 
ax.add_feature(cfeature.COASTLINE) 
ax.add_feature(cfeature.BORDERS, linestyle='-') 
for name in bird_names: 
    ix = birddata['bird_name'] == name 
    x, y = birddata.longitude[ix], birddata.latitude[ix]
    ax.plot(x, y, transform=ccrs.Geodetic(), label=name) 
plt.legend(loc='upper left')
```
