# Exploring Correlations

```python

# step 1:
corr_flavors= pd.DataFrame.corr(flavors)
corr_flavors
```

```python
# step 2 :
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.savefig("corr_flavors.pdf")
# Let's now look for the plot on the computer.
# In this case, we're getting a 12 by 12 matrix, as expected.
# Clearly, correlation of any dimension with itself is exactly plus 1
# but other strong correlations exist too.
# For example, heavy body is associated with smokiness.
# In contrast, it seems that a floral flavor is the opposite of full body
# or medicinal notes.
```

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())
plt.figure(figsize=(10,10))
plt.pcolor(corr_whisky)
plt.colorbar()
plt.savefig("corr_whisky.pdf")
```

```python
corr_whisky.shape
```

```python
corr_flavors.shape
```

# Clustering Whiskies By Flavor Profile

```python
from sklearn.cluster.bicluster import SpectralCoclustering 
```

```python
model= SpectralClustering(n_clusters=6, random_state=0, )
```

```python
model.fit(corr_whisky)
```

```python
whisky["group"]= pd.Series(model.labels_, index=whisky.index)
```

```python
whisky= whisky.ix[np.argsort(model.labels_)]
```

```python
whisky= whisky.reset_index(drop=True)
```

```python
corrolations= pd.DataFrame.corr(whisky.iloc[:,2:14].transpose())
```

```python
corrolations= np.array(corrolations)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.pcolor(corr_whisky)
plt.title("Orginal")
plt.axis("tight")

plt.subplot(122)
plt.pcolor(corrolations)
plt.title("Rearrange")
plt.axis("tight")
# plt.colorbar()
plt.savefig("corrolations.pdf")
```

```python
import numpy as np 
import pandas as pd 

whisky = pd.read_csv("whiskies.txt") 
whisky["Region"] = pd.read_csv("regions.txt") 
	# >>>whisky.head(), iloc method to index a data frame by location. 
	# >>>whisky.iloc[0:10], we specified the rows from 0 - 9 
	# >>>whisky.iloc[0:10, 0:5], we specified the rows from 0 - 9 & columns from 0-5 
	# >>>whisky.columns 
flavors = whisky.iloc[:, 2:14] 

corr_flavors = pd.DataFrame.corr(flavors) 
print(corr_flavors) 
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(10, 10)) 
plt.pcolor(corr_flavors) 
plt.colorbar() 
	#>>>plt.savefig("corlate-whisky1.pdf") 

corr_whisky = pd.DataFrame.corr(flavors.transpose()) 
plt.figure(figsize=(10, 10)) 
plt.pcolor(corr_whisky) 
plt.axis("tight") 
plt.colorbar() 
	#>>>plt.savefig("corlate-whisky2.pdf") 

plt.show() 
```

```python
from sklearn.cluster.bicluster import SpectralCoclustering 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

model = SpectralCoclustering(n_clusters=6, random_state=0) 
model.fit(corr_whisky) 
model.rows_ 
	#>>>np.sum(model.rows_, axis=1) 
	#>>>np.sum(model.rows_, axis=0) 
model.row_labels_ 
```

```python
from sklearn.cluster.bicluster import SpectralCoclustering 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

whisky['Group'] = pd.Series(model.row_labels_, index = whisky.index) 
whisky = whisky.ix[np.argsort(model.row_labels_)] 
whisky = whisky.reset_index(drop=True) 

correlations = pd.DataFrame.corr(whisky.iloc[:, 2:14].transpose()) 
correlations = np.array(correlations) 

plt.figure(figsize = (14, 7)) 
plt.subplot(121) 
plt.pcolor(corr_whisky) 
plt.title("Original") 
plt.axis("tight") 
plt.subplot(122) 
plt.pcolor(correlations) 
plt.title("Rearranged") 
plt.axis("tight") 
plt.show() 
plt.savefig("correlations.pdf") 
```
