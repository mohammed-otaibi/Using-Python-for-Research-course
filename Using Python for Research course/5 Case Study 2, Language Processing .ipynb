```python
sentence = "This comprehension check is to check for comprehension."
dic={}
spliting=[]

def remove_symbloy(sen):
    '''
    This function is for remove the symboly like ,./;:'" from any sentence
    and transfare the sentence letter to lower case.
    It is taking sentence 
    And return sentence.
    '''
    word= sen
    li=[',','.','/',':',';','""',"'"]
    for s in li:
        sen= sen.replace(s,"")
    sen= sen.lower()
    return sen

def counter_unique_word(sen):
    '''
    This function is to count all the unique word in any sentence
    It is taking sentence 
    And return dictionary that has all the word and the word count.
    '''
    dic={}
    spliting=[]
    string= remove_symbloy(sen)
    spliting= string.split(" ")
    keys= dic.keys()
    for word in spliting:
        if word in keys:
            dic[word]+=1
        else:
            dic[word]=1
    return dic
count_by_using_my_function= counter_unique_word(sentence)
```

# Computing Word Frequency Statistics

```python
def word_stats(word_counts):
    '''Return number of unique words and word frequencies.'''
    num_unique= len(word_counts)
    counts= word_counts.values()
    
    return (num_unique, counts)
```

```python
# read book by call my function in above it's name "read_book"

text= read_book("C:\\Users\\sy\\Documents\\Python Case Studies\\2 Reading book\\Romeo and Juliet in english.txt")
word_counts = counter_unique_word(text)
(num_unique, counts)= word_stats(word_counts)
print('num_unique=\t', num_unique, '\ncounts=\t', sum(counts))
```

# Reading Multiple Files

```python
import os
```

```python
book_dir= 'C:\\Users\\sy\\Documents\\Python Case Studies\\2 Reading book'
```

```python
#for print all files in the book_dir folder
os.listdir(book_dir)
```

```python
import pandas as pd

stats= pd.DataFrame(columns=("language", "author", "title", "length", "unique"))
title_num= 0
for language in os.listdir(book_dir):
    for author in os.listdir(book_dir+"/"+language):
        for title in os.listdir(book_dir+"/"+language+"/"+author):
            input_file= book_dir+"/"+language+"/"+author+"/"+title
            print(input_file)
            text= read_book(input_file)
            (num_unique, counts)= word_stats(counter_unique_word(text))
            stats.loc[title_num]= language, author.capitalize(), title.replace(".txt",""), sum(counts), num_unique
            title_num+=1
            
```
